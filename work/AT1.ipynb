# Exercício 1 - Cadastro e análise simples de veículos

Uma concessionária deseja registrar alguns veículos disponíveis para venda.
Crie um programa em Python que crie uma lista de veículos.

Cada veículo deve ser representado por um dicionário contendo:
- marca
- modelo
- preço
- ano

Exemplo de estrutura esperada:

```
veiculo = {
    "marca": "Toyota",
    "modelo": "Corolla",
    "preco": 120000,
    "ano": 2022
}
```

O programa deve:

- Adicionar pelo *menos 5 veículos* à lista.
- Exibir todos os veículos cadastrados.
- Calcular o *preço médio dos veículos*.
- Mostrar quais *veículos custam acima da média*.

### Saída esperada (exemplo)
```
Preço médio dos veículos: 85000
Veículos acima da média:
Honda Civic - 110000
Toyota Corolla - 120000
```


````python
veiculos = [
    {
        "Marca" :"Toyta",
        "Modelo":"Corolla",
        "Preco" : 120000,
        "Ano"   : 2022
    },
    {
        "Marca" :"Honda",
        "Modelo":"Civic",
        "Preco" : 130000,
        "Ano"   : 2023
    },
    {
        "Marca" :"Ford",
        "Modelo":"Focus",
        "Preco" : 90000,
        "Ano"   : 2019
    },
    {
        "Marca" :"Chevrolet",
        "Modelo":"Onix",
        "Preco" : 75000,
        "Ano"   : 2021
    },
    {
        "Marca" :"Hyundai",
        "Modelo":"HB20",
        "Preco" : 80000,
        "Ano"   : 2022
    }
]
````

````python
soma = 0

for i, v in enumerate(veiculos, start=1):
    print("Veículo:", i)
    print("Marca  :", v["Marca"])
    print("Modelo :", v["Modelo"])
    print("Preço  :", v["Preco"])
    print("Ano    :", v["Ano"])
    print("**********************")
    soma += v["Preco"]

print("\nSoma total:",soma)
media = soma/len(veiculos)
print("\nA media de valor dos veiculos é de",media,"\n")

for v in veiculos:
    if v["Preco"] > media:
        print(v["Marca"], v["Modelo"], "está acima da média")
    
````
