### Desafio - Verificar número primo

Crie um código que receba um número N do usuário e imprima uma mensagem `"Número primo"` e `"Número não primo"` conforme o teste de primalidade de N.

**Um número é primo quando é divisível somente por 1 e por ele mesmo.**

Verifique seu algoritmo com os seguintes casos:

- 2 = true
- 9 = false
- 23 = true
- 49 = false
- 100 = false
- 10007 = true
- 9999991 = true

```python
%%time
import math
n = 55645489841313564841321645646468464654
raiz_n = int(math.sqrt(n))
eh_primo = True

if n % 2 == 0:
    eh_primo = False

for i in range(3, raiz_n, 2):
    if n % i == 0:
        eh_primo = False

if eh_primo:
    print('É primo')
else:
    print('Não é primo')
```
